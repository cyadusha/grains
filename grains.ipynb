```python
import pandas as pd
import os
dir_name = '/Users/chittampalliyashaswini/Desktop/Yadu'
f = 'csv'
base_filename = "FGYearbookTable01"
p = os.path.join(dir_name, base_filename + "." + f)
grains = pd.read_csv(p)
grains.columns = grains.loc[0]
grains = grains[1:]
import string
import numpy
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.style.use('ggplot')
#import matplotlib.mlab as mlab
grains.columns.values[0] = string.split(grains.columns.values[0])[0]
grains.columns.values[1] = "Year"
grains.columns.values[6] = "Price"
grains.columns.values[2] = "Planted"
grains.columns.values[3] = "Harvested"
grains.columns.values[4] = "Production"
grains = grains.dropna(subset = ['Year'])
grains = grains[grains.Year != "2016/17"]
grains.index = range(len(grains))
grains.index += 1
grains.Commodity = grains.Commodity
grains.Year = grains.Year.str[5:]
grains['Production'] = numpy.array(grains['Production'])
def C():
    for i in range(1, len(grains) + 1, 1):
        grains['Production'][i] = grains['Production'][i].replace(",","")
        grains['Year'][i] = ["20", grains['Year'][i]]
        grains['Year'][i] = "".join(grains['Year'][i])
        
        if i in (1,8,15,22):
            grains.Commodity[i] = grains.Commodity[i]
        elif i in range(2,8,1):
            grains.Commodity[i] = grains.Commodity[1]
        elif i in range(9,15,1):
            grains.Commodity[i] = grains.Commodity[8]
        elif i in range(16,22,1):
            grains.Commodity[i] = grains.Commodity[15]
        else:
            grains.Commodity[i] = grains.Commodity[22]
    grains.Year = grains.Year.astype(int)
    return pd.DataFrame(grains.Commodity, grains.Year)

C()
x = numpy.array([[1,2,3,4]])
grains['Category'] = numpy.repeat(x,7)
grains






```

```python
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy
def E(n):
    g = grains[grains.Category == n][['Commodity','Production', 'Planted', 'Harvested','Price']]
    X = g[['Production', 'Planted', 'Harvested']]
    X["constant"] = 1
    price = g['Price']
    y = price
    grain_model = sm.OLS(y.astype(float), X.astype(float))
    results = grain_model.fit()
    fig = plt.figure()
    fig = sm.graphics.plot_ccpr_grid(results, fig=fig)
    plt.show()

E(1)
```

### In each of these plots, most of the points are close to the line of best fit. We can assume that the correlations between the price and each of the input variables are linear. 

```python
def F(n):
    g = grains[grains.Category == n][['Commodity','Production', 'Planted', 'Harvested','Price']]
    X = g[['Production', 'Planted', 'Harvested']]
    X["constant"] = 1
    price = g['Price']
    y = price
    grain_model = sm.OLS(y.astype(float), X.astype(float))
    results = grain_model.fit()
    fig = plt.figure()
    fig = sm.graphics.plot_partregress_grid(results, fig=fig)
    plt.show()
    print results.summary()
F(1)
```

### The amount of corn harvested has the highest amount of effect on the price of corn. However, the most significant variable in this regression is the amount of corn produced because it yields a p-value of 0.05. This means that even though the correlation coefficient is the lowest, there is a 95 percent chance that this variable does have an effect on the price of corn.
